# Jan 2004 Dst Index analysis

```python
import numpy as np
```

```python
data = np.loadtxt("./Data.lst")
labels = np.array(["Year", "Day", "Hour", "Scalar B (nT)", "SW Proton Density (N/cm^3)", "Dst-index (nT)"])
data = data.T
```

```python
import matplotlib.pyplot as plt

dataPints = int(data[0].shape[0])
numOfDays = int(data[1][dataPints-1] - data[1][0]) + 1
```

```python
for i in range(3, 6):
    plt.figure(figsize=(12, 3), dpi=80,)
    plt.ylabel(labels[i])
    
    plt.xticks( np.arange(0, dataPints, dataPints/numOfDays), np.arange(1,numOfDays+1) )
    if (i == 5):
        plt.xlabel( "Day in Jan {}".format(int(data[0][0])))
    
    plt.plot(data[i])

plt.show()
```

We can clearly see a drop in the disturbance storm time (Dst, Kyoto Dst) index starting roughly in jan 22nd and ending jan 25th. This means that the strength of the ring current arount the Earth caused by solar protons and electrons was specially high, affecting the field of the earth in the oposite direction, resulting in a decrease of the measured field.
A negative Dst value means that Earth's magnetic field is weakened. This is particularly the case during solar storms.
