```python
import numpy as np
```

```python
data = np.loadtxt("./Data.txt")
labels = np.array(["Year", "Day", "Hour", "Bt (nT)", "Bx (nT)", "By (nT)", "Bz (nT)", "Proton Density (N/cm^3)", "Plasma Speed (km/s)"])
data = data.T
```

| ITEMS | FORMAT |
|-------|--------|     
| YEAR | I4 |
| DOY | I4 |
| Hour | I3 |
| Scalar B, nT | F6.1 |
| BX, nT (GSE, GSM) | F6.1 |
| BY, nT (GSM) | F6.1 |
| BZ, nT (GSM) | F6.1 |
| SW Proton Density, N/cm^3 | F6.1 |
| SW Plasma Speed, km/s | F6.0 |

```python
import matplotlib.pyplot as plt

dataPints = int(data[0].shape[0])
numOfDays = int(data[1][dataPints-1] - data[1][0]) + 1
```

```python
for i in range(3, 9):
    plt.figure(figsize=(12, 3), dpi=80,)
    plt.ylabel(labels[i])
    
    plt.xticks( np.arange(0, dataPints, dataPints/numOfDays), np.arange(1,numOfDays+1) )
    if (i == 8):
        plt.xlabel( "Day in year {}".format(int(data[0][0])))
    
    plt.plot(data[i])

plt.show()
```
