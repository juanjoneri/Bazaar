```python
import numpy as np
```

```python
data = np.loadtxt("./data.txt")
labels = np.array(["Year", "Day", "Hour","Field magnitude average, nT", "BX, nT (GSE, GSM)", "BY, nT (GSM)", "BZ, nT (GSM)", "Speed (km s)", "Proton Density, n cc", "Temperature, K", "Flow pressure, nPa", "Alfven mach number", "Magnetosonic Mach number", "AE-index, nT", "AL-index, nT", "AU-index, nT", "PCN-index", "Proton flux (gt 10ev)"])
data = data.T

flag = np.array([-1.01, -1.02, -1.03, -1.04, -1.05, -1.06, -1.07, -1.08, 99999.9, 999.99, 9999999., 99.99, 999.9, 99.9, -1.015, -1.016, -1.017, 999.99, -1.019])

for j in range(3, 18):
    indices = [i for (i,v) in enumerate(data[j]) if v == flag[j]]
    data = np.delete(data, indices, 1)
```

```python
import matplotlib.pyplot as plt

dataPints = int(data[0].shape[0])
numOfDays = int(data[1][dataPints-1] - data[1][0]) + 1
```

```python
for i in range(4, 18):
    plt.figure(figsize=(20, 4), dpi=80,)
    plt.ylabel(labels[i])
    
    plt.xticks( np.arange(0, dataPints, dataPints/numOfDays), np.arange(1,numOfDays+1) )
    plt.xlabel( "Day in Feb {}".format(int(data[0][0])))
    
    plt.plot(data[i])

    plt.savefig("./Output/{}.png".format(labels[i]))

plt.show()
```
