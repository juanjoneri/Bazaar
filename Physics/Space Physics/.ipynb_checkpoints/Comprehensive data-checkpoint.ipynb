```python
import numpy as np
```

```python
data = np.loadtxt("./comprehensive_data.txt")
labels = np.array(["Year", "Day", "Hour",
                   "Scalar B, nT", 
                   "BX, nT (GSE, GSM)", 
                   "BY, nT (GSM)", 
                   "BZ, nT (GSM)", 
                   "SW Plasma Temperature, K", 
                   "SW Proton Density, N cm^3", 
                   "SW Plasma Speed, km s", 
                   "E elecrtric field", 
                   "Plasma betta", 
                   "Alfen mach number", 
                   "Magnetosonic Much num", 
                   "Kp index", 
                   "R (Sunspot No)",
                   "Dst-index, nT", 
                   "ap_index, nT", 
                   "f10.7_index", 
                   "AE-index, nT", 
                   "AL-index, nT", 
                   "AU-index, nT", 
                   "pc-index", 
                   "Lyman_alpha", 
                   "Proton flux (gt 10 Mev)"])
data = data.T
```

```python
import matplotlib.pyplot as plt

dataPints = int(data[0].shape[0])
numOfDays = int(data[1][dataPints-1] - data[1][0]) + 1
```

```python
for i in range(3, 18):
    plt.figure(figsize=(20, 4), dpi=80,)
    plt.ylabel(labels[i])
    
    plt.xticks( np.arange(0, dataPints, dataPints/numOfDays), np.arange(1,numOfDays+1) )
    plt.xlabel( "Day in Feb {}".format(int(data[0][0])))
    
    plt.plot(data[i])

    plt.savefig("./Comprehensive_Output/{}.png".format(labels[i]))

plt.show()
```
