# Numerical Solve the following equation
$$\ddot{\theta} = \omega^{2} sin(\theta) cos(\theta) - \frac{g}{R} sin(\theta)$$

1. convert the above second-order ode into two first-order ode.
2. approximate the two first-order ode by Euler's method.

```python
import numpy as np
import matplotlib.pyplot as plt
from prettytable import PrettyTable
import math
```

```python
w = .1  # d/dt angle
z = .1 # angle
dt = .05

# keep track of the values to plot
memo = np.array([(0, z, w)]) 

# make a table to display
lables = ['k', 'Z_k', 'W_k']
t = PrettyTable(lables)

for k in range(1, 900):
    newZ = w*dt + z
    newW = (2*math.sin(z)*math.cos(z)-math.sin(z))*dt + w
    z = newZ
    w = newW
    
    memo = np.append(memo ,[(k, z, w)], axis=0)
    t.add_row( [k, round(z,3), round(w,3)] )
```

```python
print(t.get_string(start = 0, end = 10))
```

```python
Ks = memo.T[0]
ThDot = memo.T[2]

plt.figure(figsize=(12, 3), dpi=80)
plt.plot(Ks, ThDot, 'go')
plt.xlabel('Steps')
plt.ylabel('Angle')
plt.show()
```

```python
Ks = memo.T[0]
Theta = memo.T[1]

plt.figure(figsize=(12, 3), dpi=80)
plt.plot(Ks, Theta, 'go')
plt.xlabel('Steps')
plt.ylabel('Angle')
plt.show()
```

```python
def frange(x, y, jump):
    while x < y:
        yield x
        x += jump

plt.figure(figsize=(12, 3), dpi=80)
plt.plot([math.sin(math.sqrt(2 - 1/2)*t) for t in frange(0,45,0.05)])
plt.xlabel('Steps')
plt.ylabel('Angle')
plt.show()
```
